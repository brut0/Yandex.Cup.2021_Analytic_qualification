```python
import numpy as np
import pandas as pd
```

```python
df_log = pd.read_csv('./feature_files/log.csv', header=None, names=['id', 'f'])
df_log
```

```python
df_log = df_log.groupby('id').count()
```

```python
clmns = [f'{i}' for i in range(1,21)]
for c in clmns:
    print(f'./feature_files/feature_{c}.csv')
    df = pd.read_csv(f'./feature_files/feature_{c}.csv', header=None, names=['id', c])
    df_log = pd.merge(df_log, df, on='id', how='left')
df_log
```

```python
def count_na(row):
    #print(row)
    full = len(clmns)
    count = 0
    for c in clmns:
        if row[c] == row[c]:
            count += 1
    return count / full
```

```python
df_log['not_na'] = df_log.apply(lambda row : count_na(row), axis = 1)
```
